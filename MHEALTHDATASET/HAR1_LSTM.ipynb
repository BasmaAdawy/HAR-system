```python
# Merge files

# List of file paths
file_paths = [file_path1, file_path2, file_path3, file_path4, file_path5,
              file_path6, file_path7, file_path8, file_path9, file_path10]

# Define column names
columns = ['user', 'acc_chest_X', 'acc_chest_Y', 'acc_chest_Z', 'electrocardio1', 'electrocardio2',
           'acc_L_ankle_X', 'acc_L_ankle_Y', 'acc_L_ankle_Z', 'gyro_L_ankle_X', 'gyro_L_ankle_Y',
           'gyro_L_ankle_Z', 'mag_L_ankle_X', 'mag_L_ankle_Y', 'mag_L_ankle_Z', 'acc_R_lower_arm_X',
           'acc_R_lower_arm_Y', 'acc_R_lower_arm_Z', 'gyro_R_lower_arm_X', 'gyro_R_lower_arm_Y',
           'gyro_R_lower_arm_Z', 'mag_R_lower_arm_X', 'mag_R_lower_arm_Y', 'mag_R_lower_arm_Z', 'Label']

def merge_files(file_paths):
    """
    Merge all log files into a single DataFrame.
    Adds a 'user' column for identifying the source subject and reorders columns.
    """
    dataframes = []
    for i, file_path in enumerate(file_paths):
        df = pd.read_csv(file_path, sep="\t", header=None)
        df.insert(0, 'user', i + 1)  # Add 'user' column as the first column
        dataframes.append(df)

    # Concatenate all dataframes
    merged_data = pd.concat(dataframes, axis=0)
    # Assign new column names
    merged_data.columns = columns
    return merged_data

# Merge the files and rename columns
merged_data = merge_files(file_paths)
print(f"Merged Data Shape: {merged_data.shape}")
```

```python
from sklearn.preprocessing import StandardScaler

# Select features and labels
def preprocess_data(data):
    """
    Select relevant features and normalize them. Separate features and labels.
    """
    # Relevant feature columns
    feature_columns = [ 'acc_R_lower_arm_X', 'acc_R_lower_arm_Y', 'acc_R_lower_arm_Z',
                       'gyro_R_lower_arm_X', 'gyro_R_lower_arm_Y', 'gyro_R_lower_arm_Z',
                        'mag_R_lower_arm_X', 'mag_R_lower_arm_Y', 'mag_R_lower_arm_Z']
    label_column = 'Label'

    # Select features and labels
    features = data[feature_columns]
    labels = data[label_column]

    # Normalize features
    scaler = StandardScaler()
    features = scaler.fit_transform(features)

    return features, labels


features, labels = preprocess_data(filtered_data)
print(f"Features Shape: {features.shape}, Labels Shape: {labels.shape}")
```

```python
# Data segmentation

import numpy as np

def segment_data(features, labels, window_size=50, overlap=25):
    """
    Segment the data into overlapping windows.

    Parameters:
        features (np.ndarray): 2D array of feature data.
        labels (pd.Series): Series of activity labels.
        window_size (int): Size of each segment in samples.
        overlap (int): Number of overlapping samples between consecutive windows.

    Returns:
        np.ndarray: Segmented feature data of shape (num_segments, window_size, num_features).
        np.ndarray: Labels for each segment.
    """
    segments, segment_labels = [], []
    step = window_size - overlap
    for i in range(0, len(features) - window_size, step):
        # Extract a window of features
        segment = features[i:i + window_size]
        # Determine the most frequent label in the corresponding window
        label = labels.iloc[i:i + window_size].mode()[0]
        segments.append(segment)
        segment_labels.append(label)
    return np.array(segments), np.array(segment_labels)

# Segment the data
window_size = 50  # 1 second at 50Hz
overlap = 25      # 50% overlap
X, y = segment_data(features, labels, window_size, overlap)

print(f"Segmented Data Shape: {X.shape}, Labels Shape: {y.shape}")

'''
Output:
X: A 3D NumPy array of shape (num_segments, window_size, num_features).
num_segments: Total number of windows created.
window_size: Number of samples per segment (50 in this case).
num_features: Number of sensor features (9 in dataset).
y: A 1D NumPy array of labels for each segment.
'''
```

```python
# oversampling

from imblearn.over_sampling import SMOTE
import numpy as np
from collections import Counter

# Check the unique classes in the dataset
print("Unique classes in the original dataset:", np.unique(y))

# Separate the majority and minority classes
X_majority = X[y != 12]  # Features for all classes except Label 12
y_majority = y[y != 12]  # Labels for all classes except Label 12

X_minority = X[y == 12]  # Features for Label 12 (minority class)
y_minority = y[y == 12]  # Labels for Label 12 (minority class)

# Check if minority class (Label 12) contains enough samples
print(f"Minority class (Label 12) sample count: {len(y_minority)}")

# Ensure that the minority class contains more than 1 sample
if len(y_minority) > 1:
    # Apply SMOTE to the combined majority and minority classes
    smote = SMOTE(sampling_strategy='auto', random_state=42)  # 'auto' will balance the class size

    # Flatten the features (required for SMOTE)
    X_combined = np.concatenate([X_majority, X_minority])
    y_combined = np.concatenate([y_majority, y_minority])

    # Apply SMOTE to the combined dataset
    X_resampled, y_resampled = smote.fit_resample(X_combined.reshape(X_combined.shape[0], -1), y_combined)

    # Check the new label distribution
    print(f"New Label Distribution: {Counter(y_resampled)}")
else:
    print("Insufficient samples in Label 12 for SMOTE. No resampling applied.")
```

```python
from sklearn.model_selection import train_test_split
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, MaxPooling1D, LSTM, Dense, Dropout

# Function to build the model
def build_model(input_shape, num_classes):
    """
    Build a CNN-LSTM model for Human Activity Recognition (HAR).
    """
    model = Sequential([
        # 1D Convolution layer
        Conv1D(64, kernel_size=3, activation='relu', input_shape=input_shape),
        # MaxPooling layer
        MaxPooling1D(pool_size=2),
        # LSTM layer (to capture time-series dependencies)
        LSTM(64, return_sequences=False),
        # Dropout layer for regularization
        Dropout(0.5),
        # Dense layer
        Dense(32, activation='relu'),
        # Output layer with softmax activation for multi-class classification
        Dense(num_classes, activation='softmax')
    ])
    return model



# Adjust labels to be zero-indexed
y_resampled = y_resampled - 1


# Convert labels to one-hot encoding
y_resampled = to_categorical(y_resampled, num_classes=len(np.unique(y_resampled)))


# Reshape X_train and X_test
time_steps = 50  # Assuming window size is 50
features_per_time_step = 9  # Number of features (e.g., accelerometer, gyroscope, magnetometer)

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=42)


# Reshape X_train and X_test to have the correct shape (samples, time_steps, features)
X_train = X_train.reshape((X_train.shape[0], time_steps, features_per_time_step))
X_test = X_test.reshape((X_test.shape[0], time_steps, features_per_time_step))

# Check the new shape of X_train and X_test
print(f"X_train shape after reshaping: {X_train.shape}")
print(f"X_test shape after reshaping: {X_test.shape}")

# Build the model
input_shape = (X_train.shape[1], X_train.shape[2])  # Time steps and features per sample
num_classes = y_train.shape[1]  # Number of output classes (one per label)

# Build the model
model = build_model(input_shape, num_classes)

# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Train the model
history = model.fit(X_train, y_train, validation_split=0.2, epochs=20, batch_size=32)


# Print model summary to check the architecture
model.summary()
```

```python
# Evaluate the model on the test set
test_loss, test_accuracy = model.evaluate(X_test, y_test, batch_size=32)

# Print the evaluation results
print(f"Test Loss: {test_loss}")
print(f"Test Accuracy: {test_accuracy}")
```

```python
from sklearn.metrics import classification_report

# Get predictions for the test set
y_pred = model.predict(X_test)

# Convert predictions from one-hot encoding back to class labels
y_pred_labels = np.argmax(y_pred, axis=1)
y_test_labels = np.argmax(y_test, axis=1)

# Generate a classification report
report = classification_report(y_test_labels, y_pred_labels)
print(report)
```
